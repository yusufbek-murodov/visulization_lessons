```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.plot(df)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
sns.lineplot(data=df)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.title("Avila Adobe muzeyiga tashrif buyurganlar")
plt.xlabel("Sana")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
sns.lineplot(data=df[['Avila Adobe', 'Chinese American Museum']])
plt.show()
```
