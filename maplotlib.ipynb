```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

Grafiklarning turli parametrlarini o'zgartirish yordamida uning tashqi ko'rinishiniham o'zgartirishimiz mumkin:

- **linewidth** - chiziq qalinligi (0.5 dan 3 gacha)
- **linestyle** - chiziq stili
  - **-** - oddiy chiziq
  - **--** - yuliq chiziq
  - **-.** - chiziq va nuqta
  - **:** - nuqtali chiziq
  - **None** yoki **' '** - chiziq yo'q
- **color** - chiziq rangi
  - **'b'** - ko'k (blue)
  - **'c'** - havorang (cyan)
  - **'g'** - yashil (green)
  - **'k'** - qora (black)
  - **'m'** - pushti (magenta)
  - **'r'** - qizil (red)
  - **'y'** - sariq (yellow)
  - Bundan tashqari ranglar HEX ('#FF5733') yoki RGBA (255,182,193,0.5) formatlarida ham berilishi mumkin.
- **alpha** - chiziq shaffofligi (0 shaffof, 0.5 yarim-shaffof, 1 shaffof emas)
- **marker** - marker
  - '.' - nuqta
  - 'o' - aylana
  - 'v' - pastga qaragan uchburchak
  - '^' - tepaga qaragan uchburchak
  - '*' - yulduz

```python
plt.plot(x, y, linewidth=2.0, linestyle='--')
plt.show()
```

```python
plt.plot(x, y,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.show()
```

```python
plt.plot(x, y,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')


# Grafik nomi:
plt.title('sin(x) funksiya grafiki')
# x - o'q nomi
plt.xlabel('pi')
# y - o'q nomi
plt.ylabel('sin(x)')
# Setka qo'shamiz
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

plt.title('sin(x) funksiya grafigi')
plt.xlabel('pi')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 6))
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x - o'q nomi:
plt.xlabel('pi')
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()
```

Oson yo'li brauzer oynasida sichqoncha o'ng tugamni bosish va Save image as matnini bosih.

yoki **plt.savefig()** funksiyasidan foydalanishimiz mumkin

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 6))

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sin(x) va cos(x) funksiya grafiki')
plt.xlabel('pi')

plt.legend(['sin(x)', 'cos[x]'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.1, linestyle='--')

plt.savefig('sin_cos.png')
```
